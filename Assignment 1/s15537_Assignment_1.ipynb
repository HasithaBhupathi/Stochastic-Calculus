# **Binomial Stock Simulation**

### **Liberies**

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
```

### **Parameters**

```python
sigma = 0.3547    #### volatility
T = 30/252       #### expiration time measured in years
r = 0.0465        #### interest rate
n = 30           #### number of steps
dt = T/n         #### one step
u = np.exp(sigma*(dt)**0.5)  #### upper rate
d = 1/u          #### lower rate
p = (np.exp(r*(T/n))-d)/(u-d)  #### upper probability
s0 = 66.09      #### start stock price
```

### **Algorithm**

```python
paths = []

for i in range(0,50):
  a = [s0]
  for j in range(0,30):
    rand = random.random()
    if rand>=p:
      a.append(a[j]*u)
    else:
      a.append(a[j]*d)
  paths.append(a)
```

### **Plot**

```python
x = []
for i in range(0,30+1):
  x.append(i)

for i in range(0,50):
  plt.plot(x,paths[i])
  plt.title("Binomial Stock Simulation")
  plt.xlabel("Time(Days)")
  plt.ylabel("Stock Price")
```
